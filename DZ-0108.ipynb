```python
# Задача 1

"""
Напишите функцию, которая классифицирует фильмы из 
материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class
"""

import pandas as pd

## Настройка источника
url = 'https://drive.google.com/file/d/1E7cEUX-UdB8T4c8rGN1esAlvaiXIuEZU/view?usp=drivesdk'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path, sep=',', skiprows =0)

## Функция классификации
def classic(param):
    if param <= 2.0:
        return 'низкий рейтинг'
    elif 4.0 >= param > 2.0:
        return 'средний рейтинг'
    elif param > 4.0:
        return 'высокий рейтинг'
    
## Добавляем столбец из функции в датафрейм
df['class'] = df['rating'].apply(classic)

## Проверка результата
df.head(10)
```

```python
# Задача 2

import pandas as pd

"""
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке
сможет выставить географическую принадлежность определенному 
региону. Т. е. если поисковый запрос содержит название города
региона, то в столбце ‘region’ пишется название этого региона.
Если поисковый запрос не содержит названия города, то ставим 
‘undefined’.

"""
## Правила распределения по условию задачи

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

## Определяем полный список городов
cities_values = sum(geo_data.values(), [])
cities_list = set(list(cities_values))

## Настройка источника
url = 'https://drive.google.com/file/d/1vX-XZsL3AHUEXuOQfZHYzs1mtH8zBdMG/view?usp=drivesdk'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df2 = pd.read_csv(path, sep=',', skiprows =0)
#df = pd.read_csv('ml-latest-small/keywords.csv')


## Функция нахождения городов в запросах
def find_city(param):
    for i in cities_list:
#        print(value)
        if i in param:
            return(i)

## Находим города в запросах
df2['city']=df2['keyword'].apply(find_city)

## Функция нахождения региона
def find_reg(param):
    for key, value in geo_data.items():
        if param in value:
            return (key)
        
## Находим регионы по городам
df2['region']=df2['city'].apply(find_reg)

## Проверка результата
filtered_countries = df2[ (df2['keyword']=='авито москва') | (df2['keyword']=='вк моя страница') | (df2['keyword']=='псков')  ]

filtered_countries
```
