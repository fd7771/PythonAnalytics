```python
# ЗАДАЧА 1
# Исходные данные
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}]


# РЕШЕНИЕ П.1 Функция p - поиска владельца документа
# Вводим номер документа

def find_doc(docs):
    """
    Функция осуществляет поиск владельца документа по номеру документа
    """
    doc_num = input('Введите номер документа: ')
    flag = 'false'
    docs=documents
    
    # Цикл поиска владельца документа
    for i in docs:
        if doc_num == i['number']:
            doc_owner = i['name']
            print('Результат: Владелец документа: '+ doc_owner)
            flag = 'true'
    
    # Проверка документа на наличие в базе по флагу    
    if flag == 'false':
        print('Результат: Документ не найден в базе')
    return('Результат: Владелец документа: ',i['name'])
    
# РЕШЕНИЕ П.2 Функция s - поиск полки документа

def find_shelf(docs, dirs):
    """
    Функция осуществляет поиск папки документа по номеру документа
    """
    doc_num = input('Введите номер документа: ')
    flag_s= 'false' 
    docs=documents
    dirs=directories

    # Цикл поиска полки документа (ищем на полках, ищем в документах, ищем введенный документ на полке, возвращаем полку)
    for shelf, content in dirs.items():   
        for i in docs:
            if doc_num in content:
                shelf_item=(shelf)
                flag_s='true'
#                print(flag_s)
    if flag_s == 'true':
        print('Результат: Документ хранится на полке: ',shelf_item)
    else:
        print('Результат: Документ не найден в базе')
    return(shelf)

# РЕШЕНИЕ П.3 Функция l - полная информация по всем документам

def find_report(dirs):
    """
    Функция осуществляет вывод отчета по документам
    """
    dirs=directories
    # Цикл поиска управленческой информации
    for shelf, content in dirs.items():
        for i in documents:
            if i['number'] in content:
                print(f"№{i['number']}, тип: {i['type']}, владелец:  {i['name']},полка хранения:  {shelf}")
    

# РЕШЕНИЕ П.4 Функция as - добавление новой полки

def create_folder(folder, dirs):
    """
    Функция осуществляет добавление новой полки
    """
        
## Вводим новую папку
    folder = input("Введите новую папку: ")
    flag= 'true'
    dirs=directories

## Цикл проверки наличия введенной папки 
    if folder in dirs:
        print(f"Результат: \nТакая полка уже существует. Текущий перечень полок:{','.join(list(dirs.keys()))}")
        flag= 'true'
    else:
        dirs[str(folder)] = []
        print (f"Результат: \nПолка добавлена. Текущий перечень полок {','.join(list(dirs.keys()))}")

# РЕШЕНИЕ П.5 Функция ds - удаление полки, на которой нет данных
def delete_folder(folder_del,dirs):
    """
    Функция осуществляет удаление полки
    """
    dirs=directories
## Вводим полку к удалению
    folder_del = input("Введите номер полки: ")

## Удаляем полку, если на ней нет данных
    
    if folder_del in dirs:
        if len(dirs[folder_del])<=0:
            del dirs[folder_del]
            print(f"Результат: \nПолка удалена. Текущий перечень полок: {','.join(list(dirs.keys()))}")
        else:
            print(f"Результат: \nНа полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {','.join(list(dirs.keys()))}")    
    else:
        print(f"Результат: \nТакой полки не существует. Текущий перечень полок: {','.join(list(dirs.keys()))}")
#   return directories['folder']                    
            
            
# ПУЛЬТ УПРАВЛЕНИЯ ПОЛЕТАМИ 
command=[]
docs=[]
dirs=[]
folder=[]
folder_del=[]

while command != 'q':
    command = input('Введите команду: ')
    if command == 's':
        find_shelf(docs,dirs)
    elif command == 'p':
#    print(doc_num )
        find_doc(docs) 
    elif command == 'l':
#    print(doc_num )
        find_report(dirs)  
    elif command == 'as':
        create_folder(folder,dirs)
    elif command == 'ds':
        delete_folder(folder_del,dirs)
    
```
