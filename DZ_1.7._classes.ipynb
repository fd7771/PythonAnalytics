```python
# Задача 1

"""
Напишите функцию, которая возвращает название валюты (поле ‘Name’) с максимальным значением курса
с помощью сервиса www.cbr-xml-daily.ru...ly_json.js
"""
def highest_rates(par_list=list_curr):
    ## Берем данные с сайта
    full_data=requests.get('https://www.cbr-xml-daily.ru/daily_json.js').json()

    ## Отбираем список валют
    list_curr = list(full_data['Valute'].keys())

    ## Создаем пустой словарь вне цикла
    dict_curr_p={}

    ## Находим значение и код валюты, создаем словарь для последующего отбора значений
    for item in list_curr:
        data_curr = full_data['Valute'][item]['Value']    # Находим значение валюты
    #    name_curr = full_data['Valute'][item]['Name']    # На текущий момент не требуется название валюты
        key_curr = full_data ['Valute'][item]['CharCode'] # Находим код валюты    
        dict_curr = {key_curr: data_curr}                 # Создаем словарь
        dict_curr_p.update(dict_curr)                     # Создаем словарь с данными вне цикла
    
    ## Находим список валют с наибольшими значениями в словаре 
    ##(количество наибольших значений задается в последних квадратных скобках)
    newA= sorted(dict_curr_p, key=dict_curr_p.get, reverse=True)[:1]

    ## Распечатываем титул результата, находим значения по наибольшим курсам 
    ## исходя из параметра newA программы (задает количество наибольших значений), распечатываем результат
    print (f'Максимальное значение курса: ')

    for value_fine in newA:
        if value_fine in full_data['Valute'].keys():      # Сверяем список отобранных валют с валютами словаря  
            values_deliv = dict_curr_p.get(value_fine)    # При совпадении проставляем значения
            print(f'{value_fine} - {values_deliv}')       # Печатаем результат

highest_rates()
```

```python
# Задача 2

# Импорт библиотеки запросов
import requests

"""
Добавьте в класс Rate параметр diff (со значениями True или False),
который в случае значения True в методах курсов валют (eur, usd итд) будет возвращать не курс валюты, 
а изменение по сравнению в прошлым значением. Считайте, self.diff будет принимать значение True только при возврате значения курса. 
При отображении всей информации о валюте он не используется.
"""

## Создаем класс Rate
class Rate:
    def __init__(self, format_='value'):
        self.format = format_
    def exchange_rates(self):                           ### Метод для получения всех данных по всем валютам            
        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']
        
    def exchange_rate_by_currency(self, currency, diff): ### Метод для получения данных по отдельным валютам
        self.currency= currency                          ### Описание валюты
        response = self.exchange_rates()                 ### Определение словаря
        if currency in response:                         ### Если валюта в словаре            
            if self.format == 'value':                   ### Если формат value
                if diff == 'true':                   ### Если параметр истинен то отображать разницу между тек и прошл.
                    res_diff = response[currency]['Value'] - response[currency]['Previous']
                    return print(f'Изменение курса валюты {self.currency} составило: {res_diff}')
                elif diff == 'false':                ### Если параметр ложен то отображать курс валюты
                    res = response[currency]['Value']                    
                    return print(f'Курс валюты {self.currency} составил: {res}')                   
                else:
                    return 'Error'
        return 'Error'

r = Rate()
 
r.exchange_rate_by_currency(currency='EUR', diff='true') ### Проведение тестирования с параметром истина
r.exchange_rate_by_currency(currency='EUR', diff='false')### Проведение тестирования с параметром ложь
```
